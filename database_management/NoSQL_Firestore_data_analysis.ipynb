### 2) Create the Firestore collections and subcollections based on your data model and populate them with the Shopify records.

```python
# importing necessary libraries
from google.cloud import firestore
import pandas as pd
db = firestore.Client()
```

```python
# Creating top-level collection for Apps COLLECTION (Access pattern #2)
batch = db.batch()

# Loading in Apps data
df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()

# creating Apps Collection
for row in rows: 
    record = {}
    record['id'] = row[0]
    record['url'] = row[1]
    record['title'] = row[2]
    record['developer'] = row[3]
    record['developer_link'] = row[4]
    record['icon']= row[5]
    record['rating']= row[6]
    record['review_count']= row[7]

    app_ref = db.collection('apps').document(row[0])
    batch.set(app_ref, record)

# committing the batch 
batch.commit()
```

```python
# Creating Category COLLECTION, Apps SUBCOLLECTION (Access pattern #1)
batch = db.batch()

# Category COLLECTION
category_df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/categories.csv', sep=',', header=0, lineterminator='\n')
category_rows = category_df.values.tolist()

# Apps SUBCOLLECTION 
apps_df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = apps_df.values.tolist()

# For app_id
apps_categories_df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/apps_categories.csv', sep=',', header=0, lineterminator='\n')
apps_categories_rows = apps_categories_df.values.tolist()

# Creating Category Collection and Apps Subcollection
for category_row in category_rows:
    
    category_ref = db.collection('categories').document(str(category_row[0]))
    
    category_record = {
        'title': category_row[1]
    }
    
    batch.set(category_ref, category_record)
    
    for apps_categories_row in apps_categories_rows:
        if category_row[0] == apps_categories_row[1]:
            
            for apps_row in apps_rows:
                if apps_categories_row[0] == apps_row[0]:
                    apps_ref = category_ref.collection('apps').document(str(apps_row[0]))
                    
                    apps_record = {
                        'url': apps_row[0],
                        'title': apps_row[2],
                        'developer': apps_row[3],
                        'developer_link': apps_row[4],
                        'icon': apps_row[5],
                        'rating': apps_row[6],
                        'review_count': apps_row[7]
                    }
                    
                    batch.set(apps_ref, apps_record)
    
    # committing the batch 
    batch.commit()
```

```python
# Creating Apps COLLECTION, Pricing_Plans SUBCOLLECTION (Access pattern #3)
batch = db.batch()

# Apps COLLECTION
df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
app_rows = df.values.tolist()

# Pricing_Plans SUBCOLLECTION
df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/pricing_plans.csv', sep=',', header=0, lineterminator='\n')
pricing_plan_rows = df.values.tolist()

# Creating Apps Collection and Pricing Plans Subcollection
for app_row in app_rows:
    app_record = {}
    app_record['id'] = app_row[0]
    app_record['url'] = app_row[1]
    app_record['title'] = app_row[2]
    app_record['developer'] = app_row[3]
    app_record['developer_link'] = app_row[4]
    app_record['icon'] = app_row[5]
    app_record['rating'] = app_row[6]
    app_record['review_count'] = app_row[7]
    
    app_ref = db.collection('apps').document(app_row[0])
    
    batch.set(app_ref, app_record)
    
    for pricing_plan_row in pricing_plan_rows:
        if app_row[0] == pricing_plan_row[1]:
            pricing_plan_record = {}
            pricing_plan_record['id'] = pricing_plan_row[0]
            pricing_plan_record['title'] = pricing_plan_row[2]
            pricing_plan_record['price'] = pricing_plan_row[3]
    
            pricing_plan_ref = app_ref.collection('pricing_plans').document(pricing_plan_row[0])
    
            batch.set(pricing_plan_ref, pricing_plan_record)
        
    # committing the batch
    batch.commit()
```

```python
# Creating Apps COLLECTION, Key_Benefits SUBCOLLECTION (Access pattern #4)
batch = db.batch()

# Apps COLLECTION - Loading data
df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
app_rows = df.values.tolist()

# Key_Benefits SUBCOLLECTION - Loading data
df = pd.read_csv('/home/jupyter/firestore_proj5/fs_shopify/key_benefits.csv', sep=',', header=0, lineterminator='\n')
key_benefits_rows = df.values.tolist()

# Creating the Apps Collection & Key Benefits Subcollection
for app_row in app_rows:
    app_record = {}
    app_record['id'] = app_row[0]
    app_record['url'] = app_row[1]
    app_record['title'] = app_row[2]
    app_record['developer'] = app_row[3]
    app_record['developer_link'] = app_row[4]
    app_record['icon'] = app_row[5]
    app_record['rating'] = app_row[6]
    app_record['review_count'] = app_row[7]
    
    app_ref = db.collection('apps').document(app_row[0])
    
    batch.set(app_ref, app_record)
    
    for key_benefits_row in key_benefits_rows:
        if app_row[0] == key_benefits_row[0]:
            title = key_benefits_row[1]
            key_benefits_record = {}
            key_benefits_record['title'] = title
            key_benefits_record['description'] = key_benefits_row[2]
            
            # to resolve error with "/"
            if '/' in title:
                title = title.replace('/','_')
            
            key_benefits_ref = app_ref.collection('key_benefits').document(title)
        
            batch.set(key_benefits_ref, key_benefits_record)
    
    # committing the batch w/ each document added 
    batch.commit()
```

### Retrieve the number of documents in each collection and subcollection using count().See code sample read_count.py.

```python
from google.cloud.firestore_v1 import aggregation
```

```python
# getting count of documents for Apps, Category COLLECTIONS
collection_refs = ['apps', 'categories']
for collection_ref in collection_refs:
    ref = db.collection(collection_ref)
    count_query = ref.count()
    query_result = count_query.get()
    
    print('Number of docs in', collection_ref, 'COLLECTION:', query_result[0][0].value)
```

```python
# getting count of documents for key_benefits & pricing_plans SUBCOLLECTION under Apps COLLECTION
def count_key_benefits_pricing_plans_docs():
    apps_subcollection_refs = ['key_benefits', 'pricing_plans']
    
    for subcollection_ref in apps_subcollection_refs:
        
        doc_count = 0
        apps_ref = db.collection('apps')
        apps = apps_ref.stream()
        
        for app in apps:
            ref = apps_ref.document(app.id).collection(subcollection_ref)
            count_query = ref.count()
            query_result = count_query.get()
            
            if len(query_result) > 0:
                value = query_result[0][0].value
                doc_count += value
        
        print('Number of documents in', subcollection_ref, 'SUBCOLLECTION under Apps COLLECTION:', doc_count)
        
count_key_benefits_pricing_plans_docs()
```

```python
# getting count of documents for apps SUBCOLLECTION
def count_apps_subcollection_docs():
    apps_subcollection_count = 0
    category_ref = db.collection('categories')
    categories = category_ref.stream()
    
    for category in categories:
        apps_subcollection_ref = category_ref.document(category.id).collection('apps')
        count_query = apps_subcollection_ref.count()
        query_result = count_query.get()
        
        # if there's any documents in the subcollection
        if len(query_result) > 0:
            value = query_result[0][0].value
            apps_subcollection_count += value
    
    print('Number of documents in Apps SUBCOLLECTION under Categories COLLECTION:', apps_subcollection_count)
    
count_apps_subcollection_docs()
```

### List the top 10 "Productivity" apps (whose categories.title = "Productivity") sorted by their rating in descending order.
* Return id,  title, developer, rating and reviews_count for those apps
* Limit the results to the first 10 records

```python
from google.cloud.firestore_v1.base_query import FieldFilter
import pandas as pd
# Initialize a list to store the top 10 "Productivity" apps
top_10_productivity_apps = []
category_ref = db.collection('categories')
query = category_ref.where(filter=FieldFilter('title', '==', 'Productivity'))
category_snapshots = query.stream()

top_10_productivity_apps = []

for category_snapshot in category_snapshots:
    apps_ref = category_snapshot.reference.collection('apps')
    query = apps_ref.order_by('rating', direction=firestore.Query.DESCENDING).limit(10)
    app_snapshots = query.stream()

    for app_snapshot in app_snapshots:
        app_data = app_snapshot.to_dict()
        top_10_productivity_apps.append({
            'id': app_snapshot.id,
            'title': app_data['title'],
            'developer': app_data['developer'],
            'rating': app_data['rating'],
            'reviews_count': app_data['review_count']
        })

# Create a DataFrame from the list
top_10_productivity_apps_df = pd.DataFrame(top_10_productivity_apps)

# Print the DataFrame with the top 10 "Productivity" apps
print(top_10_productivity_apps_df)
```

### List the 10 apps with the highest number of reviews (based on apps.review_count). 
* Return id, title, developer, rating and reviews_count for those apps.
* Order the results by reviews_count in descending order

```python
# Query for the top 10 apps with the highest number of reviews
# Reference to the "apps" collection
apps_ref = db.collection('apps')
query = apps_ref.order_by('review_count', direction=firestore.Query.DESCENDING).limit(10)
results = query.stream()
# Initialize a list to store the top 10 apps
top_10_apps = []
# Iterate through the results and store app information in the list
for app in results:
    app_data = app.to_dict()
    top_10_apps.append({
        'id': app.id,
        'title': app_data['title'],
        'developer': app_data['developer'],
        'rating': app_data['rating'],
        'reviews_count': app_data['review_count']
    })
# Create a DataFrame from the list 
top_10_apps_df = pd.DataFrame(top_10_apps)
# Print the DataFrame with the top 10 apps
print(top_10_apps_df)
```
